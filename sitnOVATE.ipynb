```python
import speech_recognition as sr

def speech_to_text(audio_data):
    """Converts audio data to text using Google Speech Recognition."""
    recognizer = sr.Recognizer()
    try:
        print("Recognizing...")
        text = recognizer.recognize_google(audio_data) # Use Google Speech Recognition
        print(f"STT Output: {text}")
        return text
    except sr.UnknownValueError:
        print("Could not understand audio")
        return ""
    except sr.RequestError as e:
        print(f"Could not request results from Google Speech Recognition service; {e}")
        return ""

if __name__ == '__main__':
    # First, capture audio (using the capture_audio function from the previous step)
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Speak now...")
        audio = recognizer.listen(source)
        print("Audio captured.")

    if audio: # If audio was captured successfully
        converted_text = speech_to_text(audio) # Convert the captured audio to text
        if converted_text:
            print(f"Final STT Result: {converted_text}")
        else:
            print("Speech-to-text conversion failed.")
    else:
        print("No audio data to process.")
```
